#### Let's Find the Causes of High Cancelation Rate 

#### adr (average daily rate)

```python
adr_resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
adr_city_hotel = city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
#line plot to show monthly ADR
plt.figure(figsize=(12,4))
plt.title('Average Daily Rate in City hotel and resort hotel', size=10)
plt.plot(adr_resort_hotel.index, adr_resort_hotel['adr'], label='Resort Hotel')
plt.plot(adr_city_hotel.index, adr_city_hotel['adr'], label='City Hotel')
legend_labels,_ = ax.get_legend_handles_labels()
ax.legend(bbox_to_anchor=(1,1))
plt.legend(['resort_hotel', 'city_hotel'])
```

- There are spike, it can be because hotel price will be higher on weekend.
- we can see that higher then city hotel during 7th to 9th month of each year
- There are also some spike like 1st month of 2016 for Resort hotel, 2nd or 3rd month of 2016 for City hotel and 1st month of 2017 for resort hotel.

##### Let's Find Cancelation of reservation in different momths

```python
#Extract month from reservation_status_date column
hb['month'] = hb['reservation_status_date'].dt.month

plt.figure(figsize=(12,5))
ax = sns.countplot(x='month', hue='is_canceled', data=hb, palette='Blues')
legend_labels,_ = ax.get_legend_handles_labels()
ax.legend(bbox_to_anchor=(1,1))
plt.title('Reservation Status in diff. month', size=12)
plt.xlabel('hotel')
plt.ylabel('No. of Reservations')
plt.legend(['not cancled', 'cancled'])
plt.show()
```

- There is higher cancelation in January and Lowest in August
- There are Highest Reservation in August and lowest in December and January
- The Cancelations are gradually reduced from January to June
- it is little bit confusing that when there is highest reservation in August then there is least cencelation and when there is approx. lowest reservation in January there is highest cencelation.
- it could be possible that there is low price in August, so there is low cancelation and Higher price in January, so cancelation is High
- **LET'S PLOT the ADR for each month so, we can find out the causes of cancelation rate**

```python
monthly_adr = hb[hb['is_canceled']==1].groupby('month')[['adr']].sum().reset_index()
plt.figure(figsize=(8,6))
plt.title('Avg. Daily Rate per Month')
sns.barplot(x='month', y='adr', data = monthly_adr)
```

- We can see that There is Highest Average price in January, and also cancelation is also higher in January
- In August there is the lowest cancelation and there is also average price is approx. lower level
- **Thus We can conclude that higher the Price highere the cancelation and lower the price lower the cancelation.**

```python
#Let's Check the Average Price (adr) for Canceled Reservation and for Not Canceled Reservation 

#Finding Adr for Canceled Reservation
canceled_hb_adr = canceled_hb.groupby('reservation_status_date')[['adr']].mean().reset_index()
canceled_hb_adr.sort_values('reservation_status_date', inplace=True)

#Finding ADR for Not-Canceled Reservation
not_canceled_hb = hb[hb['is_canceled']==0]
not_canceled_hb_adr = not_canceled_hb.groupby('reservation_status_date')[['adr']].mean().reset_index()
not_canceled_hb_adr.sort_values('reservation_status_date', inplace=True)

#Ploting ADR for canceled and not canceled reservation
plt.figure(figsize=(16,6))
plt.title('Average Daily Rate For Canceled and Not-Canceled Reservation')
plt.plot(not_canceled_hb_adr['reservation_status_date'], not_canceled_hb_adr['adr'], label='not cancelled')
plt.plot(canceled_hb_adr['reservation_status_date'], canceled_hb_adr['adr'], label='cancelled')
plt.legend()
plt.show()
```

- Here, clearly we can see that Average Daily Price for cancelled Reservation is Higher then that of Not-Cancelled Reservation
- **Thus one of the causes of reservation cancellation could be Higher Price**
